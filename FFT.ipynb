```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1024
T = 1.0/44100.0
f1 = 440
f2 = 880
t = np.linspace(0.0, N*T, N)
y1 = np.sin(2 * np.pi * f1 * t)
y2 = np.sin(2 * np.pi * f2 * t)
y = y1+y2
print(t.shape)
print(y1.shape)

plt.subplot(3,1,1)
plt.plot(t,y1)
plt.subplot(3,1,2)
plt.plot(t,y2)
plt.subplot(3,1,3)
plt.plot(t,y)
```

```python
y3 = np.hstack([y, y, y])
plt.subplot(211)
plt.plot(y3)
plt.axvspan(N, N*2, alpha=0.3)
plt.xlim(0, 3*N)

plt.subplot(212)
plt.plot(y3)
plt.axvspan(N,N*2, alpha=0.3)
plt.xlim(900, 1270)
plt.show()
```

```python
from scipy.fftpack import fft
yf = fft(y, N)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
plt.stem(xf, 2.0/N* np.abs(yf[0:N//2]))
plt.xlim(0,2000)
plt.show()
```

```python
import scipy as sp
f, P = sp.signal.periodogram(y, 44100, nfft=2**12)
plt.subplot(211)
plt.plot(f, P)
plt.xlim(100, 1900)
```

```python
import librosa
import librosa.display

D = np.abs(librosa.stft(y))
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='linear', x_axis='time')
plt.title('Octave')
plt.show()
```
